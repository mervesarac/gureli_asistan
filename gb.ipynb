```python
import pandas as pd
```

```python
df = pd.read_excel('gb_subat.xlsx', parse_dates=['Tarih'])
df.head()
```

```python
df.describe(include='all')
```

```python
servis_adi_distribution = df['Servis Adı'].value_counts(normalize=True) * 100
ax = servis_adi_distribution.plot(kind='bar', title='Servis Adı Distribution (%)', ylabel='Percentage', xlabel='Servis Adı')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}%', (p.get_x() + p.get_width() / 2., p.get_height() + 0.5), 
                ha='center', va='bottom', fontsize=8, rotation=90)
```

```python
queue_desc_distribution = df['Queue Desc'].value_counts(normalize=True) * 100
ax = queue_desc_distribution.plot(kind='bar', title='Queue Desc Distribution (%)', ylabel='Percentage', xlabel='Queue Desc')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}%', (p.get_x() + p.get_width() / 2., p.get_height() + 0.5), 
                ha='center', va='bottom', fontsize=8, rotation=90)
```
