```python
import pandas as pd
```

```python
df = pd.read_excel('crm_subat.xlsx', parse_dates=['Datetime'])
df.head()
```

```python
df.describe(include='all')
```

```python
ax = df['Class'].value_counts(normalize=True).mul(100).plot(kind='bar', title='Class Percent Distribution', xlabel='Class', ylabel='Percentage')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=8)
```

```python
print(df['Class'].unique())
```

```python
import matplotlib.pyplot as plt

ax = df['Channel'].value_counts(normalize=True).mul(100).plot(kind='bar', title='Channel Percent Distribution', xlabel='Channel', ylabel='Percentage')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
```

```python
ax = df['Status'].value_counts(normalize=True).mul(100).plot(kind='bar', title='Status Percent Distribution', xlabel='Status', ylabel='Percentage')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
```

```python
ax = df['Type'].value_counts(normalize=True).mul(100).plot(kind='bar', title='Type Percent Distribution', xlabel='Type', ylabel='Percentage')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
```

```python
print(df['Type'].unique())
```

```python
df['Class - Type'] = df['Class'] + ' - ' + df['Type']
print(df[['Class', 'Type', 'Class - Type']].head())
```

```python
print(df['Class - Type'].describe())
```

```python
ax = df['Class - Type'].value_counts(normalize=True).mul(100).plot(kind='bar', title='Class - Type Percent Distribution', xlabel='Class - Type', ylabel='Percentage', figsize=(12, 6))
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=8)
```

```python
import sqlite3

# SQLite veritabanı bağlantısı oluştur
conn = sqlite3.connect('crm_subat.db')

# DataFrame'i SQLite tablosu olarak kaydet
df.to_sql('crm_subat', conn, if_exists='replace', index=False)

# Bağlantıyı kapat
conn.close()
```
